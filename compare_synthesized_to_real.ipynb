```python
import numpy as np
import matplotlib.pyplot as plt
from  create_training_data.training_data_generator import ClassifierValidationDataGenerator, ClassifierTrainingDataGenerator

# initialise generators
real_dots_gen = ClassifierValidationDataGenerator()
synth_dots_gen = ClassifierTrainingDataGenerator.make_default_training_data_generator(10000)

# generate images
real_dots,real_tags = real_dots_gen.get_next_batch()
synth_dots,synth_tags = synth_dots_gen.get_next_batch()
synth_dots = [synth_dots[int(i)] for i in np.linspace(0,synth_dots_gen.batch_size - 1,len(real_dots))]
synth_tags = [synth_tags[int(i)] for i in np.linspace(0,synth_dots_gen.batch_size - 1,len(real_dots))]
```

```python
for b in range(100):
    i = np.random.randint(0,len(synth_tags))
    if synth_tags[i][0] == 0:
        continue
    ch_index = np.argmax(synth_tags[i][1:])
    plt.figure()
    plt.imshow(np.max(synth_dots[i][...,ch_index],axis=-1))
```
